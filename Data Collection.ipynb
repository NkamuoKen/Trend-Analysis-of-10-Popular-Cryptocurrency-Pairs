```python
json_content_13 = response_13.json()#BTC
json_content_14 = response_14.json()#ETH
json_content_15 = response_15.json()#LTC
json_content_16 = response_16.json()#XRP
json_content_17 = response_17.json()#XMR
json_content_18 = response_18.json()#ADA
json_content_19 = response_19.json()#BNB
json_content_20 = response_20.json()#BCH
json_content_21 = response_21.json()#XLM
json_content_22 = response_22.json()#DOGE
```

```python
import pandas as pd
```

```python
# Create a list for the json contents of the responses and their also a list for the pairs symbols
json_contents = [json_content_13, json_content_14, json_content_15, json_content_16, json_content_17, json_content_18, 
                json_content_19, json_content_20, json_content_21, json_content_22]
currency_pairs = ['BTC/USD', 'ETH/USD', 'LTC/USD', 'XRP/USD', 'XMR/USD', 'ADA/USD', 'BNB/USD', 'BCH/USD', 'XLM/USD', 'DOGE/USD']
```

```python
# Create a for loop for us to convert the json contents and currency pair list into a pandas dataframe
dfs = []

for json_content, currency_pair in zip(json_contents, currency_pairs):
    json_data = json_content['Data']['Data']
    df = pd.DataFrame(json_data)
    df['currency_pair'] = currency_pair
    dfs.append(df)
```

```python
result_df = pd.concat(dfs, ignore_index=True)
```

```python
result_df
```

```python
result_df.to_csv('crypto_data.csv', index=False)
```
