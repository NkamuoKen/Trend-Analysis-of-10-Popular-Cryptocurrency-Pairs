```python
# Let us check the descriptive statistics for the BTC/USD
btc_data = df[df['Currency Pair'] == 'BTC/USD']
btc_stats = btc_data.describe()
btc_stats
```

We will be focusing mainly in the statistics of the Close because that is what actually matters at the end of the day, and the dataset we have is based on the daily price of 10 popular cryptocurrency

```python
# Let us check if the BTC data is skewed
plt.boxplot(btc_data['Close'], vert=False, showfliers=False)
plt.title('Box Plot for BTC/USD')
plt.xlabel('Closing Price')
plt.show()
```

Since we are having a postive skewed dataset for the BTC/USD we can say that there have been an uptrend for the BTC/USD starting from 2018-2023. This can continue with this uptrend with the recent news going on in the cryptocurrency world. From the descriptive statistics above we can also see the average performance of the BTC/USD is 22400.541099 in as much as there are huge outliers but it shows significant surge in the price of BTC/USD due to factors like events, news, and other things. The median is the 50th percentile which is 19230.960000. Since skewness have affected our data we can now rely on the median to measure the performance of other cryptocurrency pairs with BTC/USD, then we will use the min, max and standard deviation to measure the difference volatility of different cryptocurrency pairs.

All of the cryptocurrency pairs are positively skewed whch signifies an uptrend of all of them between 2018-2023. So we have check the volatility in each pairs using the Range which is the difference between the max and min. In other words Range = Volatility

```python
Range_of_btc = btc_data['Close'].max() - btc_data['Close'].min()
Range_of_eth = eth_data['Close'].max() - eth_data['Close'].min()
Range_of_ltc = ltc_data['Close'].max() - ltc_data['Close'].min()
Range_of_xrp = xrp_data['Close'].max() - xrp_data['Close'].min()
Range_of_xmr = xmr_data['Close'].max() - xmr_data['Close'].min()
Range_of_ada = ada_data['Close'].max() - ada_data['Close'].min()
Range_of_bnb = bnb_data['Close'].max() - bnb_data['Close'].min()
Range_of_bch = bch_data['Close'].max() - bch_data['Close'].min()
Range_of_xlm = xlm_data['Close'].max() - xlm_data['Close'].min()
Range_of_doge = doge_data['Close'].max() - doge_data['Close'].min()

Ranges_of_currency_pairs = [Range_of_btc, Range_of_eth, Range_of_ltc, Range_of_xrp, Range_of_xmr, Range_of_ada, 
                    Range_of_bnb, Range_of_bch, Range_of_xlm, Range_of_doge]
```

```python
for currency_pairs in Ranges_of_currency_pairs:
    print(currency_pairs)
```

```python
Ranges_of_currency_pairs.sort()
```

```python
Ranges_of_currency_pairs
```

So from the above list it is obvious that BTC/USD has the highest volatility followed by ETH/USD, then BCH/USD. This means for traders or investors who have more risk appetite should go for more volatile pairs like the BTC/USD, ETH/USD, and BCH/USD, while for the ones who have less risk appetite should go for pairs like DOGE/USD, XLM/USD, and XRP/USD.
